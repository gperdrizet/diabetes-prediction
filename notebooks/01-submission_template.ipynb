# Diabetes prediction challenge: submission template

This notebook is a guide and temlate for Kaggle comptition submissions. For the Diabetes prediction challenge, submissions are made via Kaggle notebook - you must upload and run your notebook on Kaggle.

Upload a notebook to Kaggle via the '+ Create' option in the navidation menu at the left (note: you must have an account and be logged in). Choose 'Import Notebook' and then either upload as a file or link to GitHub/Colab etc.

Once in the notebook environment, add the competition data source via the 'Input' tab in the sidebar on the right. Click '+ Add Input', select the 'Competition Datasets' filter and find 'Diabetes Prediction Challenge'. Click the '+' icon on the dataset to add it to your notebook environment. Note: you will need to join the competition to add the dataset. Find the 'Diabetes Prediction Challenge' on the competitions tab from the Navigation menu on the left of any Kaggle page. Then click the 'Join Competition' button at the upper right.
